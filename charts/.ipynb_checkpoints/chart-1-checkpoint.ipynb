```python
SCOPE = "engineering_only"   # "engineering_only", "eng_and_tech", or "custom_set"
CUSTOM_CAH = ("10", "11", "13")  # used only if SCOPE == "custom_set"
TOP_N = 8                     # top subjects (by latest year) to show; keeps lines readable
OUT_BASENAME = "eng_students_lines"  # files saved as .svg and .png
TITLE = "UK Engineering Students by CAH Level 3 Subject"
YLABEL = "Enrolments"
path = r"../data/table-52" 
```

```python
df = load_df(path, skip_rows=13)
```

```python
def publication_rc():
    plt.rcParams.update({
        "figure.figsize": (12, 6),
        "axes.titlesize": 14,
        "axes.labelsize": 12,
        "xtick.labelsize": 10,
        "ytick.labelsize": 10,
        "legend.fontsize": 10,
        "lines.linewidth": 2.0,
        "savefig.dpi": 300
    })
```

```python
def keep_totals_slice(df):
    mask = (
        df["CAH level marker"].eq("CAH level 3") &
        df["Entrant marker"].eq("All") &
        df["Level of study"].eq("All") &
        df["Mode of study"].eq("All") &
        df["Permanent address marker"].eq("Total")
    )
    d = df.loc[mask].copy()
    # # parse year start (e.g., "2019/20" -> 2019) for chronological sort if needed
    # d["YearStart"] = d["Academic Year"].str.slice(0, 4).astype(int)
    return d
```

```python

def filter_scope(d, custom=("10","11","13")):
    keep = d["CAH level subject"].str.startswith(custom, na=False)
    df_keep = d[keep].copy()
    return df_keep
```

```python
def ensure_subject_label_has_no_code(d):
    if "Subject" not in d.columns:
        d["Subject"] = d["CAH level subject"].str.replace(
            r"^\s*\d{2}-\d{2}-\d{2}\s+", "", regex=True
        ).str.strip()
    return d
```

```python
def build_pivot(d):
    # Pivot Academic Year x Subject
    pvt = d.pivot_table(index="Academic Year", columns="Subject",
                        values="Number", aggfunc="sum").fillna(0)
    # ensure chronological order by the leading year
    order = (pvt.index.str.slice(0, 4).astype(int)).argsort()
    pvt = pvt.iloc[order]
    return pvt
```

```python
def select_top_columns(pvt, top_n=8):
    latest_year = pvt.index.max()
    top_cols = pvt.loc[latest_year].sort_values(ascending=False).head(top_n).index
    pvt_top = pvt[top_cols].copy()
    # add a “Total” line = all columns (not just top ones) for context
    # pvt_top["Total (all subjects)"] = pvt.sum(axis=1)
    return pvt_top
```

```python
def plot_lines(pvt_top, title, ylabel, out_basename):
    publication_rc()
    ax = pvt_top.plot()  # one plot, multiple lines
    ax.set_title(title)
    ax.set_xlabel("Academic Year")
    ax.set_ylabel(ylabel)
    ax.grid(True, which="both", linestyle="--", alpha=0.4)  # gridlines for readability
    ax.legend(loc="upper left", bbox_to_anchor=(1.02, 1), frameon=False)
    plt.tight_layout()
    plt.savefig(f"{out_basename}.svg")  # vector for journals
    plt.savefig(f"{out_basename}.png")  # raster for quick use
    plt.show()
```

```python
d = keep_totals_slice(df)
d = filter_scope(d, custom=CUSTOM_CAH)
d = ensure_subject_label_has_no_code(d)
TOP_N = d["Subject"].nunique()
pvt = build_pivot(d)
pvt_top = select_top_columns(pvt, top_n=TOP_N)
```

```python
scope_title = {
    "engineering_only": " (Engineering only)",
    "eng_and_tech": " (Engineering & Technology)",
    "custom_set": f" (CAH {CUSTOM_CAH})"
}[SCOPE]

plot_lines(pvt_top, TITLE + scope_title, YLABEL, OUT_BASENAME)
```
